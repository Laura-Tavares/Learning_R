# Pie charts

## Syntax used

*pie(x, labels, radius, main, col, clockwise)*

## Parameters used 

- **x** is a vector containing the numeric values used in the pie chart.

- **labels** is used to give description to the slices.

- **radius** indicates the radius of the circle of the pie chart.(value between −1 and +1).

- **main** indicates the title of the chart.

- **col** indicates the color palette.

- **clockwise** is a logical value indicating if the slices are drawn clockwise or anti clockwise.

```r
slices  <- c(60,20,10,10)

labels  <- c ("London","Paris","Orlando","Miami")

pct  <- round (slices/sum(slices)*100)

labels  <- paste (labels,pct,"%",sep=" ") 

pie(slices,radius= 0.8 ,labels = labels, clockwise=TRUE, col=rainbow(length(labels)),
   main="Travelling places")

legend("topleft", c("London","Paris","Orlando","Miami"), cex = 1.0,
   fill = rainbow(length(labels)))
```

## Using ggplot

Ggplot2 does not have a specific geometric function to build pie charts. **The key is to go back to geom_bar( ) and add the polar coordinate function to make the graph circular.** There are ways to enhance the pie chart but we will keep it to a minimum here. Leave the x in aesthetics blank with just the quotation marks. The pie chart will be drawn in the counterclockwise motion, alphabetically.

### Modify components of a theme - ggplot
https://ggplot2.tidyverse.org/reference/theme.html

### Color palette - ggplot
![image](https://r-graph-gallery.com/38-rcolorbrewers-palettes_files/figure-html/thecode-1.png)

```r
# First try

df <- data.frame(
  group = c("Killua", "Gon", "Hisoka"),
  favorite = c(30, 10, 60)
  )
head(df)

library (ggplot2)
pie <- ggplot(data = df, aes(x = "", y = favorite, fill = group)) + 
  geom_bar(stat = "identity", color="black") + 
  labs(title = "Most favorite characters on HxH") +
  theme(plot.title = element_text(size=40)) +
  coord_polar("y") +
  theme_void()
pie + scale_fill_brewer(palette="Set3")
```

```r
# Second try with personalized titles
options(repr.plot.width = 12, repr.plot.height = 6, repr.plot.res = 100)

mynamestheme <- theme(
  title = element_text(size = (40)),
  legend.title = element_text(colour="black", family = "Helvetica", size=(25)), 
  legend.text = element_text(colour = "black", family = "Helvetica", size=(20))
    )

df <- data.frame(
  Characters = c("Killua", "Kurapika", "Hisoka"),
  favorite = c(30, 10, 60)
  )
head(df)

library (ggplot2)
pie <- ggplot(data = df, aes(x = "", y = favorite, fill = Characters)) + 
  geom_bar(stat = "identity", color="black") + 
  labs(title = "Most favorite characters on HxH") +
  theme(title = element_text(size=40)) +
  coord_polar("y") +
  theme_void()

pie + scale_fill_brewer(palette="Set3") + mynamestheme
```
