# Bar Charts 

A bar chart represents data in rectangular bars with length of the bar proportional to the value of the variable. R uses the function barplot() to create bar charts. R can draw both vertical and Horizontal bars in the bar chart. In bar chart each of the bars can be given different colors.

## Syntax used
*barplot(H,xlab,ylab,main, names.arg,col)*

## Parameters used 
- **H** is a vector or matrix containing numeric values used in bar chart.
- **xlab** is the label for x axis.
- **ylab** is the label for y axis.
- **main** is the title of the bar chart.
- **names.arg** is a vector of names appearing under each bar.
- **col** is used to give colors to the bars in the graph.

```r
H  <- c (20,30,50,15)
barplot (H)
```

## Bar Chart Labels, Title and Colors

The features of the bar chart can be expanded by adding more parameters. 

- The **main** parameter is used to add title. 
- The **col** parameter is used to add colors to the bars. 
- The **args.name** is a vector having same number of values as the input vector to describe the meaning of each bar.

```r
V  <- c (20,10,60,50)
A  <- c ("HxH", "Dororo", "CodeGeass", "Naruto")
barplot (H,names.arg=A, xlab="Anime", ylab="Favoritism", col="purple", main="Famous animes", border='black' )
```

## Group Bar Chart and Stacked Bar Chart

We can create bar chart with groups of bars and stacks in each bar by using a matrix as input values.

More than two variables are represented as a matrix which is used to create the group bar chart and stacked bar chart.

```r
colors  <- c("green","yellow","red")
months  <- c("Janurary", "Feburary", "March", "April", "June")
regions <- c("East","West","North")
values <- matrix(c(10,2,3,11,9,4,8,7,3,12,5,10,8,10,11), nrow = 3, ncol = 5, byrow = TRUE)

barplot (values, main="Rain in the country", names.arg = months, xlab = "Months", ylab = "Rain levels", col = colors)

legend ("topleft", regions, cex=2.0, fill=colors)
```

```r
colors  <- c("green","yellow","red")
months  <- c("Janurary", "Feburary", "March", "April", "June")
regions <- c("East","West","North")
values <- matrix(c(10,2,3,11,9,4,8,7,3,12,5,10,8,10,11), nrow = 3, ncol = 5, byrow = TRUE)

barplot (values, main="Rain in the country", names.arg = months, xlab = "Months", ylab = "Rain levels", col = colors, horiz=TRUE)

legend ("bottomright", regions, cex=1.8, fill=colors)
```

```r
df  <- data.frame (
    PIB  <- c (19,11,5,4,3,2,1),
    Countries  <- c ("USA", "China", "Japan", "Germany", "England", "France","India")
)

barplot (df$PIB,
        main= "Top 7 PIB worldwide",
        xlab="Countries",
        ylab="PIB",
        names.arg=df$PIB,
        border="black",
        col="pink")
```

## Using ggplot

The geometric function called **geom_bar( )** is used to draw a bar graph in ggplot2. Axes labels are automatically added and graphs are arranged in alphabetical order.

The heights of the bars commonly represent either a count of cases in each group or the values in a column of the data frame. By default, geom_bar uses stat = “bin.” That means that the height of each bar equals the number of cases in each group. In our case, we want the heights of the bars to represent values in Users. Therefore, we use stat = “identity” and map a value to the y aesthetic.

The default for geom_bar is a vertical barplot. To do a horizontal barplot, we flip the coordinates by adding the commnand **cood_flip( )** as follows.

```r
library (ggplot2)

df  <- data.frame (
    V  <- c (20,10,60,50),
    A  <- c ("HxH", "Dororo", "CodeGeass", "Naruto")
)

ggplot(data=df, aes(x=A,y=V)) + geom_bar (stat = "identity")
```

```r
library (ggplot2)

df  <- data.frame (
    V  <- c (20,10,60,50),
    A  <- c ("HxH", "Dororo", "CodeGeass", "Naruto")
)

ggplot(data=df, aes(x=A,y=V)) + geom_bar (stat = "identity") + coord_flip( )
```

### Enhancements in Basic R - ggplot

To add a title, labels on the axes and color to your bar graph, we use the following arguments:

- **main** = “Header of the graph”
- **xlab** = “x-axis label”
- **ylab** = “y-axis label”
- **name.arg** = vector (used for labelling each of the bar graphs)
- **border** = “bar graph border color”
- **col** = “color to fill bar graph”

```r
library(ggplot2)

df  <- data.frame (
    PIB  <- c (19,11,5,4,3,2,1),
    Countries  <- c ("USA", "China", "Japan", "Germany", "England", "France","India")
)

ggplot (data=df, aes(x=Countries, y=PIB, fill=Countries)) +
    geom_bar (stat = "identity")
```

```r
# Order + adding labels

library(ggplot2)

df  <- data.frame (
    PIB  <- c (19,11,5,4,3,2,1),
    Countries  <- c ("USA", "China", "Japan", "Germany", "England", "France","India")
)

ggplot (data=df, aes(x = reorder (Countries, - PIB) , y=PIB, fill=Countries)) +
    geom_bar (stat = "identity") +
    labs (title = "Top 7 PIB worldwide",
         x = "Countries", y = "PIB")
```
