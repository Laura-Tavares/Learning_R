# Histogram
*Observation:* A bar graph is used to compare discrete or categorical variables in a graphocal format whereas a hostogram depicts the frequency distribuition of variables in a dataset.

## Syntax
***hist**(v,main,xlab,xlim,ylim,breaks,col,border)*

## Parameters
- **v** is a vector containing numeric values used in histogram.

- **main** indicates title of the chart.

- **col** is used to set color of the bars.

- **border** is used to set border color of each bar.

- **xlab** is used to give description of x-axis.

- **xlim** is used to specify the range of values on the x-axis.

- **ylim** is used to specify the range of values on the y-axis.

- **breaks** is used to mention the width of each bar.

```r
# Example

vector  <- c (2,3,6,2,8,5,7,7,3,4.5,4,5,2)

hist (vector , xlab = "Grades", col = "lightpink", border = "purple")
```

#### Interval notation explained of the example above:

- 2 ≤ x ≥ 3
- 3 < x ≥ 4
- 4 < x ≥ 5 
- 5 < x ≥ 6
- 6 < x ≥ 7
- 7 < x ≥ 8

### Specifying the range values in axis
To specify the range of values allowed in X axis and Y axis, we can use the **xlim** and **ylim** parameters.

```r
v <- c(9,13,21,8,36,22,12,31,33,19,39,27,2,5,0)

hist(v,xlab = "Weight",col = "maroon",border = "pink", xlim = c(0,40), ylim = c(0,5), xaxp=c(0,40,8),
   breaks = 10)
```
